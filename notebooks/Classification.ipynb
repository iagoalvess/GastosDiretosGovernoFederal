```python
# Load dataset
gastos_diretos = pd.read_csv('../databases/gastosdiretos.csv', encoding='utf-8')
```

## Distribuição simples dos orçamentos realizados

Dado essa distribuição dos incidentes, vamos classificar os estados em 3 conjuntos:

0-6: Low

6-12: Medium

12+ : High

Em seguida, tentaremos classificar usando KNN os estados pelo número de leis de controle de armas que possuém.

```python
plt.figure(figsize=(12, 6))
gastos_diretos['log_orcamento_realizado'] = np.log1p(gastos_diretos['ORÇAMENTO REALIZADO (R$)'])
sns.histplot(gastos_diretos['log_orcamento_realizado'], bins=20, kde=True)
plt.title('Distribuição Logarítmica dos Orçamentos Realizados')
plt.xlabel('Log do Orçamento Realizado (R$)')
plt.ylabel('Frequência')
plt.savefig('../resultados/dist_orcamentos.png', bbox_inches='tight')
plt.show()
```
